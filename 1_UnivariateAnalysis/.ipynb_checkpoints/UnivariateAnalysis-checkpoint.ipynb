```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../Datasets/Titanic-Dataset.csv')
df.head(3)
```

## Univariate Analysis on Categorical Data

#### Count Plot

```python
sns.countplot(data=df, x='Survived', palette='viridis')
plt.show()
```

```python
df['Embarked'].value_counts().plot(kind='bar')
plt.show()
```

#### Pie Chart

```python
plt.pie(x = df['Sex'].value_counts(), autopct='%.2f')
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
plt.show()
```

## Univariate Analysis on Numerical Data

#### Histogram

```python
plt.hist(df['Age'], bins=20, density=True)
plt.show()
```

#### Distplot

```python
sns.distplot(df['Age'], kde=True, bins=20)
plt.show()
```

#### Boxplot

```python
# IQR = Interquartile Range
# Q1 = 25 percentile, Q3 = 75 percentile
# Minimum = Q1 - 1.5*IQR, Maximum = Q3 + 1.5*IQR

sns.boxplot(x=df['Age'], fill=True)
plt.show()
```
